```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
datasets = pd.read_excel('./assets/datasets_for_svm.xlsx', sheet_name="Sheet2")
datasets_train = datasets.iloc[:,:6]
datasets_train.info()
datasets_train.head()
```

```python
## features for model : mean, std, skew, kurt, mean crossing

# mean_first = np.mean(datasets_train['upward_shift'])
# std_first = np.std(datasets_train['upward_shift'])

df = pd.DataFrame(columns=['mean', 'std', 'skew', 'kurt'])
df['mean'] = np.mean(datasets_train)
df['std'] = np.std(datasets_train)
df['skew'] = datasets_train.skew(axis=0)
df['kurt'] = datasets_train.kurt(axis=0)
df['label'] = ['4','2','3','5','6','7']
df
```

```python
X = np.asarray(df[['mean', 'std', 'skew', 'kurt']])
y = np.asarray(df['label'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
classifier = svm.SVC(kernel='linear', gamma='auto', C=2)
classifier.fit(X_train, y_train)
y_predict = classifier.predict(X_test)
print(y_test, y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```

```python
ex = pd.read_csv('./assets/synthetic_control_data.csv', delim_whitespace=True, header=None)
ex.head()
```

```python
%%capture --no-display
xdf = pd.DataFrame(columns=['mean', 'std', 'skew', 'kurt', 'label'])
xdf['mean'] = np.mean(ex, axis=1)
xdf['std'] = np.std(ex, axis=1)
xdf['skew'] = ex.skew(axis=1)
xdf['kurt'] = ex.kurt(axis=1)
xdf['label'].iloc[0:100] = 'normal'
xdf['label'].iloc[100:201] = 'cyclic'
xdf['label'].iloc[201:301] = 'increasing_trend'
xdf['label'].iloc[301:401] = 'decreasing_trend'
xdf['label'].iloc[401:501] = 'upward_shift'
xdf['label'].iloc[501:] = 'downward_shift'
```

```python
X = np.asarray(xdf[['mean', 'std', 'skew', 'kurt']])
y = np.asarray(xdf['label'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
classifier = svm.SVC(kernel='linear', gamma='auto', C=2)
classifier.fit(X_train, y_train)
y_predict = classifier.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict))
```
